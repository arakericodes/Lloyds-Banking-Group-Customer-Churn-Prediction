```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
df5['Gender'] = l.fit_transform(df5['Gender'])
df5['ResolutionStatus'] = l.fit_transform(df5['ResolutionStatus'])
```

```python
nomial_f = ['ServiceUsage','InteractionType', 'ProductCategory', 'MaritalStatus']
numerical_f = ['LoginFrequency', 'AmountSpent', 'Age']
ordinal_f = ['IncomeLevel']
```

```python
X = df5.drop('ChurnStatus', axis=1)
Y = df5['ChurnStatus']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y, test_size=0.30, random_state=32)
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
o = OneHotEncoder(drop='first')
ord = OrdinalEncoder()
s = StandardScaler()
ct = ColumnTransformer([
    ('ohe', o, nomial_f),
    ('s', s, numerical_f),
    ('ord', ord, ordinal_f)
], remainder='passthrough')
```

```python
xtrain = ct.fit_transform(xtrain)
```

```python
xtest = ct.transform(xtest)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(xtrain, ytrain)
```

```python
from sklearn.model_selection import GridSearchCV
params = {
    'criterion':['gini', 'entropy', 'log_loss'],
    'max_features':['sqrt', 'log2']
}
grid = GridSearchCV(RandomForestClassifier(), param_grid=params, verbose=1)
```

```python
grid.fit(xtrain, ytrain)
```

```python
grid.best_params_
```

```python
ypred = grid.predict(xtest)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred))
```
